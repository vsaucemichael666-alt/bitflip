```python
import ipywidgets as widgets
from IPython.display import display

def flip_bit_ipywidget(button):

    current_value = button.description
    if current_value == "0":
        button.description = "1"
        button.style.button_color = '#2ECC71'
    else:
        button.description = "0"
        button.style.button_color = '#E74C3C'

def get_current_binary_code_ipywidget(buttons):

    binary_string = ""
    for btn in buttons:
        binary_string += btn.description


    spaced_binary_string = " ".join([binary_string[i:i+8] for i in range(0, len(binary_string), 8)])

    return spaced_binary_string

def print_to_output_ipywidget(buttons, output_widget):

    final_code = get_current_binary_code_ipywidget(buttons)

    with output_widget:

        output_widget.clear_output()
        print("\n" + "="*40)
        print(f"  Ваш измененный двоичный код:")
        print(f"  {final_code}")
        print("="*40 + "\n")


initial_code_input = input("Введите ваш двоичный код ")

initial_code = initial_code_input.replace(" ", "")


if not all(c in '01' for c in initial_code) or not initial_code:
    print("Ошибка")
else:

    print(f"кнопки для кода: {initial_code}...")

    bit_buttons_ipy = []

    byte_hbox_list = []


    for i in range(0, len(initial_code), 8):
        byte_segment = initial_code[i:i+8]
        temp_button_list_for_byte = []
        for bit in byte_segment:
            color = '#2ECC71' if bit == "1" else '#E74C3C'

            btn = widgets.Button(
                description=bit,
                disabled=False,
                button_style='',
                tooltip='Нажмите чтобы изменить бит',
                layout=widgets.Layout(width='50px', height='50px'),
                style=widgets.ButtonStyle(button_color=color, font_weight='bold')
            )

            btn.on_click(flip_bit_ipywidget)

            temp_button_list_for_byte.append(btn)
            bit_buttons_ipy.append(btn)

        byte_hbox = widgets.HBox(temp_button_list_for_byte)
        byte_hbox_list.append(byte_hbox)

    output_area = widgets.Output()

    print_button_ipy = widgets.Button(
        description="Вывести код",
        disabled=False,
        button_style='primary',
        tooltip='Вывести текущий двоичный код',
        layout=widgets.Layout(width='auto')
    )

    print_button_ipy.on_click(lambda b: print_to_output_ipywidget(bit_buttons_ipy, output_area))

    display(widgets.VBox([widgets.Label("Нажимайте на биты чтобы их изменить:")] + byte_hbox_list + [print_button_ipy, output_area]))
```
